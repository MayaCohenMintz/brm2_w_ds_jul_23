```python
def add_ap(df, ap_size):
    o = df[["ox", "oy", "oz"]].values
    u = df[["ux", "uy", "uz"]].values
    v = df[["vx", "vy", "vz"]].values

    center = np.array([0.5, 0.5, 1])
    matrix = np.stack([u, v, o], axis=1)
    atlas_center_coords = center @ matrix
    df["ap"] = ap_size - atlas_center_coords[:, 1]
```

```python
add_ap(model.predictions)
model.predictions
```

```python
ds_paper_alignments = [pd.read_csv(filename) for filename in Path("/home/ben/python/brm2/data/deepslice_raw/DeepSlice_Alignments/MENS-AI-CI").glob("*.csv")]
ds_paper_alignments = pd.concat(ds_paper_alignments)
ds_paper_alignments["Filenames"] = ds_paper_alignments["Filenames"].apply(lambda x: os.path.basename(x))
ds_paper_alignments = ds_paper_alignments.set_index("Filenames")
add_ap(ds_paper_alignments)
ds_paper_alignments
```

```python
from pathlib import Path
import cv2
import numpy as np
from skimage import exposure

image_list = list(Path("/home/ben/brainways/brainways-reg-model/data/real/test/images").rglob("*.tif"))

for i in image_list:
    # read the image in int16 format
    img = cv2.imread(str(i), cv2.IMREAD_UNCHANGED)
    
    # rescale the image intensities
    v_min, v_max = np.percentile(img, (0, 98))
    rescaled_img = exposure.rescale_intensity(img, in_range=(v_min, v_max))
    
    # convert the rescaled image to uint8 format
    img_uint8 = cv2.convertScaleAbs(rescaled_img, alpha=255.0)
    
    # save the image in .png format
    cv2.imwrite(f"/tmp/rat-test/{i.with_suffix('.png').name}", img_uint8)
```

```python
model.predict("/tmp/rat-test", ensemble=False, section_numbers=False)
```
